```python
import qiskit
```

```python
from qiskit import QuantumCircuit
```

```python
q=QuantumCircuit(1,1)
```

```python
q.draw()
```

```python
q.h(0)
```

```python
q.measure(0,0)
```

```python
q.draw()
```

```python
import matplotlib.pyplot as plt
```

```python
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
simulator=AerSimulator()
```

```python
n=int(input("Enter number of shots : "))
```

```python
result = simulator.run(q, shots=n).result()
```

```python
counts = result.get_counts()
```

```python
p=plot_histogram(counts)
```

```python
x=p.axes[0]
```

```python
x.set_xticklabels(["Tails (0)","Heads (1)"])
```

```python
display(p)
```
