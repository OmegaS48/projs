```python
import qiskit
```

```python
from qiskit import QuantumCircuit
```

```python
n=16 #(Number of bits your number will have, 16 is set by default. Therefore highest possible number generated is 65535)
```

```python
q = QuantumCircuit(n, n)
```

```python
for i in range(n):
    q.h(i)
```

```python
from qiskit_aer import AerSimulator
```

```python
aers=AerSimulator()
```

```python
q.measure(range(n), range(n))
```

```python
job=aers.run(q, shots=1)
```

```python
result = job.result()
```

```python
counts = result.get_counts()
```

```python
binnumber= list(counts.keys())[0]
```

```python
print("Random number in Binary:", binnumber)
print("Random number in Decimal:", int(binnumber,2))
```
